## data preprocessing, load data and transform them into NN-usable form

```python
#resized
import numpy as np
from os import listdir
from PIL import Image
from matplotlib import image
from numpy import asarray

def image_process_for_nn(path):
    data_list = []
    for filename in listdir(path):
        if filename == ".DS_Store": continue
        # load image
        img = Image.open(path + '/' + filename)
        data = asarray(img)
        # Normalize the images.
        data = (data / 255) - 0.5
        # Flatten the images.
        data = data.reshape((-1, 10000,3))
        #concat those data into a list 
        data_list.append(data)
        print(filename +" processed")
    return data_list
        
```

```python
data_cat_train_list = image_process_for_nn('training_set/resized cats')
```

```python
data_dog_train_list = image_process_for_nn('training_set/resized dogs')
```

```python
data_cat_test_list = image_process_for_nn('test_set/resized cats')
```

```python
data_dog_test_list = image_process_for_nn('test_set/resized dogs')
```

```python
def add_label(data_list,label):
    a_list_of_lists = []
    for i in range(len(data_list)):
        a_list_of_lists.append([data_list[i],label])
    return a_list_of_lists
```

```python
data_cat_train_label_list = add_label(data_cat_train_list,1) # cat label 1,dog label 0
```

```python
data_dog_train_label_list = add_label(data_dog_train_list,0) 
```

```python
data_cat_test_label_list = add_label(data_cat_test_list,1)
```

```python
data_dog_test_label_list = add_label(data_dog_test_list,0) 
```

```python
data_train_label_list = data_cat_train_label_list + data_dog_train_label_list
data_test_label_list = data_cat_test_label_list + data_dog_test_label_list 
```

```python
def from_list_to_array(data_list,number_of_data): # listarray, number of data = how many data instances u want
    data_array = data_list[0][0]
    data_label_list = [data_list[0][1]]
    for i in range(1,number_of_data):
        data_array = np.append(data_array,data_list[i][0],axis=0)
        print("Current array shape :" + str(data_array.shape))
        data_label_list.append(data_list[i][1])
    return data_array,data_label_list
```

```python
%load_ext autoreload
%autoreload 2
import os
import sys
import time
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from Shapley import ShapNN
from DShap import DShap
from shap_utils import *
%matplotlib inline
MEM_DIR = './'
```

```python
problem, model = 'classification', 'logistic'
hidden_units = [] # Empty list in the case of logistic regression.
train_size = 100
```

```python
d, difficulty = 50, 1
num_classes = 2
tol = 0.03
target_accuracy = 0.7
important_dims = 5
clf = return_model(model, solver='liblinear', hidden_units=tuple(hidden_units))
_param = 1.0
for _ in range(100):
    X_raw = np.random.multivariate_normal(mean=np.zeros(d), cov = np.eye(d), 
                                          size=train_size + 5000)
    _, y_raw, _, _ = label_generator(
        problem, X_raw, param = _param,  difficulty = difficulty, important=important_dims)
    clf.fit(X_raw[:train_size], y_raw[:train_size])
    test_acc = clf.score(X_raw[train_size:], y_raw[train_size:])
    if test_acc>target_accuracy:
        break
    _param *= 1.1
print('Performance using the whole training set = {0:.2f}'.format(test_acc))
```

```python
X, y = X_raw[:train_size], y_raw[:train_size]
X_test, y_test = X_raw[train_size:], y_raw[train_size:]
model = 'logistic'
problem = 'classification'
num_test = 1000
directory = './temp'

sources = None
```
